```python
import re
import enchant
import pickle
import pandas as pd
```

```python
dictionary_FR = enchant.Dict("fr_FR")
dictionary_GB = enchant.Dict("en_GB")
dictionary_US = enchant.Dict("en_US")
```

```python
def isFrenchWord(word):
    return dictionary_FR.check(word)
```

```python
def isEnglighWord(word):
    return dictionary_GB.check(word) or dictionary_US.check(word)
```

```python
# isFrenchWord('maison')
# True
```

```python
def shouldWeTranslate(texte):
    words = texte.split()
    for word in words:
        if word.isalpha() and isFrenchWord(word) and not isEnglighWord(word):
            return True
    return False
```

```python
# shouldWeTranslate("house 12px  ")
# False
# shouldWeTranslate("maison 12px  ")
# True
```

```python
dic_uniques = pickle.load( open( "_data/dic_uniques.pickle", "rb" ) )
dic_to_translate = dict()
```

```python
idx = 0
for (key, texte) in dic_uniques.items() :
    if shouldWeTranslate(texte):
        dic_to_translate[key] = texte
```
