```python
import pickle
from collections import OrderedDict
import pandas as pd
```

```python
translated_list = pickle.load( open( "_data/translated_list.pickle", "rb" ) )
```

```python
list_uniques = []
```

```python
for value in translated_list:
    texts = value.split('\n')
    for text in texts:
        list_uniques.append(text.strip())
```

```python
list_uniques
```

```python
len(list_uniques)
```

```python
df = pd.DataFrame.from_dict(list_uniques)
```
