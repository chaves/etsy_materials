```python
import pickle
from collections import OrderedDict
import pandas as pd
```

```python
ETSY_ROUND = '_2019'
OUTPUT_FOLDER = ETSY_ROUND + '/output/'
```

```python
translated_list = pickle.load( open( OUTPUT_FOLDER + "translated_list.pickle", "rb" ) )
```

```python
list_uniques = []
```

```python
for value in translated_list:
    texts = value.split('\n')
    for text in texts:
        list_uniques.append(text.strip())
```

```python
list_uniques
```

```python
len(list_uniques)
```

```python
df = pd.DataFrame.from_dict(list_uniques)
```

```python
df.to_excel(OUTPUT_FOLDER + 'uniques_translated.xlsx')
```
